```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")


train_data = train_data.dropna(axis=1)

test_data = test_data.dropna(axis=1)
numerical_features = train_data.select_dtypes(include=[np.number])
non_numeric_features = train_data.select_dtypes(exclude=[np.number])

Y_train = numerical_features["SalePrice"].values
numerical_features = numerical_features.drop("SalePrice", axis=1)

numerical_features = (numerical_features - numerical_features.mean()) / numerical_features.std()

non_numeric_features_encoded = pd.get_dummies(non_numeric_features)

X_train = pd.concat([numerical_features, non_numeric_features_encoded], axis=1)

X_train = np.c_[np.ones(X_train.shape[0]), X_train]

def compute_cost(X, Y, theta):
    m = len(Y)
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - Y))
    return cost

def gradient_descent(X, Y, theta, learning_rate, iterations):
    m = len(Y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        error = predictions - Y
        gradient = (1 / m) * X.T.dot(error)
        theta = theta - learning_rate * gradient
        cost_history[i] = compute_cost(X, Y, theta)

    return theta, cost_history

def stochastic_gradient_descent(X, Y, theta, learning_rate, iterations):
    m = len(Y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        for j in range(m):
            rand_ind = np.random.randint(0, m)
            X_i = X[rand_ind, :].reshape(1, -1)
            Y_i = Y[rand_ind]
            error = X_i.dot(theta) - Y_i
            gradient = X_i.T.dot(error)
            theta = theta - learning_rate * gradient
        cost_history[i] = compute_cost(X, Y, theta)

    return theta, cost_history

def mini_batch_gradient_descent(X, Y, theta, learning_rate, iterations, batch_size):
    m = len(Y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        for j in range(0, m, batch_size):
            X_i = X[j:j+batch_size]
            Y_i = Y[j:j+batch_size]
            gradient = (1 / batch_size) * X_i.T.dot(X_i.dot(theta) - Y_i)
            theta = theta - learning_rate * gradient
        cost_history[i] = compute_cost(X, Y, theta)

    return theta, cost_history

theta = np.zeros(X_train.shape[1])
learning_rate = 0.01  # Reduced learning rate
iterations = 1000
batch_size = 32


batch_theta, batch_cost = gradient_descent(X_train, Y_train, theta, learning_rate, iterations)
print("Batch Gradient Descent")
print("Optimized Parameters (Theta):", batch_theta)
print("Final Cost:", batch_cost[-1])

stochastic_theta, stochastic_cost = stochastic_gradient_descent(X_train, Y_train, theta, learning_rate, iterations)
print("\nStochastic Gradient Descent")
print("Optimized Parameters (Theta):", stochastic_theta)
print("Final Cost:", stochastic_cost[-1])

mini_batch_theta, mini_batch_cost = mini_batch_gradient_descent(X_train, Y_train, theta, learning_rate, iterations, batch_size)
print("\nMini-Batch Gradient Descent")
print("Optimized Parameters (Theta):", mini_batch_theta)
print("Final Cost:", mini_batch_cost[-1])
```
